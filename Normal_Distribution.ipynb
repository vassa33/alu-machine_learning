# **Normal distribution**

It is also known as **Gaussian distribution**, and is characterized by a bell-shaped curve. ðŸ””




## The **key properties** of the normal distribution include:


> **Symmetry:** The curve is symmetric around the mean, with half of the data points falling to the left and half to the right of the mean.

> **Bell-Shaped Curve:** The curve is bell-shaped, with the highest point (mode) at the mean, and it gradually tapers off on both sides.

> **Mean and Standard Deviation:** The distribution is fully described by two parameters: the mean (Î¼) and the standard deviation (Ïƒ). The mean determines the center of the distribution, while the standard deviation determines the spread or dispersion of the data.

### **Example** of a Normal Distribution curve:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Define parameters
mu = 0  # Mean
sigma = 1  # Standard Deviation

# Generate a set of random data points that follow a normal distribution
data = np.random.normal(mu, sigma, 1000)

# Sample of our random data:-
print(data[0:10])
```

```python
# Create a histogram to visualize the distribution
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.grid(True)
plt.show()
```

## **Probability distribution functions (PDFs)**

Probability distribution functions (PDFs) are mathematical functions that describe the probability distribution of a random variable. They provide a way to calculate the probability of different outcomes or events.

There are a variety of different PDFs, but for our example above, we will use a Probability Density Function - PDF (yes, another PDF ðŸ¤“) because of the continuous random nature of the normal distribution variable.

**The Probability Density Function:-**

Where:


> f(x) is the probability density at x.

> Î¼ is the mean (average) of the distribution.

> Ïƒ is the standard deviation, which controls the spread of the distribution.

> e is the base of the natural logarithm.

> Ï€ is the mathematical constant pi (approximately 3.14159).


The PDF describes the relative likelihood (probability) of different values in a continuous normal distribution.

### **Example using the above formula**

```python
#create an array of 100 equally spaced values
x = np.linspace(-3, 3, 100)

# calculate the PDF values for the normal distribution at each x-coordinate
pdf = norm.pdf(x, mu, sigma)

# Plot the probability density function (PDF)
plt.plot(x, pdf, 'r-', lw=2)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')

# Show the plot
plt.grid(True)
plt.show()
```

## **Interpreting the Plot:**

The plot shows the shape of the normal distribution's PDF.

> The highest point on the curve corresponds to the mean (Î¼) of the distribution.

> As you move away from the mean in both directions, the probability density gradually decreases, forming the characteristic bell-shaped curve.

> The area under the entire PDF curve sums to 1, representing the total probability of observing any value within the distribution.
